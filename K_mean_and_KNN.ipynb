```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
scaler = MinMaxScaler()
scaler.fit(df[['sepal length (cm)']])
df['sepal length (cm)'] = scaler.transform(df[['sepal length (cm)']])
scaler = MinMaxScaler()
scaler.fit(df[['sepal width (cm)']])
df['sepal width (cm)'] = scaler.transform(df[['sepal width (cm)']])
```

```python
plt.scatter(df['petal length (cm)'],df['petal width (cm)'])
```

```python
plt.scatter(df['sepal length (cm)'],df['sepal width (cm)'])
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3)
```

```python
df['cluster'] = km.fit_predict(df[['sepal length (cm)','sepal width (cm)']])
```

```python
km.cluster_centers_
```

```python
df0 = df[df.cluster==0]
df1 = df[df.cluster==1]
df2 = df[df.cluster==2]
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='yellow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],marker='*')
```

```python
inertias = []

for i in range(1,5):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(df[['sepal length (cm)','sepal width (cm)']])
    inertias.append(kmeans.inertia_)

plt.plot(range(1,5), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```
